```python
import pandas as pd
df = pd.read_excel('Data.xlsx', sheet_name = 'stock data', header = None)
```

```python
#row with ticker in it
ticker_row = df.iloc[2,:].dropna()
ticker_row_new = ticker_row.copy();
```

```python
date_idx = []
price_idx = []
turn_idx = []
mkt_idx = []
pricedf = pd.DataFrame()
turndf = pd.DataFrame()
mktdf = pd.DataFrame()

# Select the column with full date range with another column to form a dataframe
# This dataframe is used for joining later
complete_df = pd.DataFrame({'date':df.iloc[3:,3486], 'temp_col':df.iloc[3:,3487]})

# First column is date
pricedf[0] = df.iloc[3:,3486];
turndf[0] = df.iloc[3:,3486];
mktdf[0] = df.iloc[3:,3486];

for i in range(df.shape[1]):
    if (i+1)%7 == 1:
        date_idx.append(i);
        
    elif (i+1)%7 == 2:
        price_idx.append(i);
        cur_date_col = df.iloc[3:, i-1];
        cur_price_col = df.iloc[3:, i];
        temp_df_price = pd.DataFrame({'date': cur_date_col,'price': cur_price_col})
        result = pd.merge(complete_df, temp_df_price, how = 'left', on = 'date')
        full_date_price = result['price']
        # n = # of stock
        n = int(((i+1) - 2)/7)
        pricedf[n+1] = full_date_price
                    
    elif (i+1)%7 == 3:      
        turn_idx.append(i);
        cur_date_col = df.iloc[3:, i-2];
        cur_turn_col = df.iloc[3:, i];
        temp_df_turn = pd.DataFrame({'date': cur_date_col,'turnover': cur_turn_col})
        result = pd.merge(complete_df, temp_df_turn, how = 'left', on = 'date')
        full_date_turn = result['turnover']
        # n = # of stock
        n = int(((i+1) - 3)/7)
        turndf[n+1] = full_date_turn
        
    elif (i+1)%7 == 4:      
        mkt_idx.append(i);
        cur_date_col = df.iloc[3:, i-3];
        cur_mkt_col = df.iloc[3:, i];
        temp_df_mkt = pd.DataFrame({'date': cur_date_col,'mv': cur_mkt_col})
        result = pd.merge(complete_df, temp_df_mkt, how = 'left', on = 'date')
        full_date_mv = result['mv']
        # n = # of stock
        n = int(((i+1) - 4)/7)
        mktdf[n+1] = full_date_mv
        
pricedf = pricedf.reset_index(drop=True)
pricedf = pricedf.dropna(how='all')
turndf = turndf.reset_index(drop=True)
turndf = turndf.dropna(how='all')
mktdf = mktdf.reset_index(drop=True)
mktdf = mktdf.dropna(how='all')
```
